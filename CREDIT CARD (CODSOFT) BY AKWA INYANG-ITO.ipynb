# UNDERSAMPLING METHOD

Carrying out undersampling
To build a sample  dataset with similar distribution of the fradulent transaction and the legitimate transaction

Number of Fradulent transaction = 492

```python
LegitSample = Legit.sample(n=492)
```

Concatenating the two dataframes 

```python
New_df3 = pd.concat([LegitSample,Fraud], axis=0)
```

```python
New_df3.head()
```

```python
New_df3.tail()
```

```python
#Checking distribution of legitimate transaction and fradulent transaction in the new sample datset
New_df3['Class'].value_counts()
```

```python
#Comparing values of both transaction in the sample dataset
New_df3.groupby('Class').mean()
```

Splitting dataset into target variable and feature variables 

```python
X =New_df3.drop(columns = 'Class', axis = 0)
Y = New_df3['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Model Training

# Using Logistic Regression

```python
#Using Logistics Regression
model = LogisticRegression()
```

```python
# Training the Logistics Regression model with training data
model.fit(X_train,Y_train)
```

Evaluation of the model 

Accuracy score

```python
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data:  ', training_data_accuracy )
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
#Checking the precision score for log(train)
precision_score(X_train_prediction,Y_train)
```

```python
#Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data:  ', test_data_accuracy )
```

```python
#Checking the precision score for Log.Regression(test)
precision_score(X_test_prediction,Y_test)
```
